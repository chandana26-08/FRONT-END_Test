```python
import pandas as pd
import nltk
import numpy as np
```

```python
df = pd.read_csv(
    '/content/IMDB Dataset.csv',
    engine='python',
    encoding='latin-1',
    on_bad_lines='skip'   # ✅ This is the key fix
)

print("✅ Dataset loaded successfully!")
print(df.shape)
df.head()
```

```python
df.head()
```

```python
!ls /content
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
ps = PorterStemmer()
```

```python
def preprocess_text(text):
  words= word_tokenize(text.lower())
  sw=stopwords.words('english')
  words=[i for i in words if i.isalnum() and i not in sw]
  words=[ps.stem(i) for i in words]
  return ' '.join(words)
```

```python
df['preprocessed_review']=df['review'].apply(preprocess_text)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features=5000)
```

```python
x = cv.fit_transform(df['preprocessed_review']).toarray()
```

```python
y = df['sentiment'].map({'positive':1,'negative':0})
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
IMDB=MultinomialNB()
IMDB
```

```python
IMDB.fit(xtrain,ytrain)
```

```python
ypred=IMDB.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
ac=accuracy_score(ytest,ypred)
```

```python
ac
```

```python
report=classification_report(ytest,ypred)
```

```python
report
```

```python
new_review=''' The film is a worthwhile watch. On a scale from zero to five, I give this film a five," or a more detailed critique that discusses specific aspects like the acting, direction, cinematography, and themes'''
```

```python
processed_new_review=preprocess_text(new_review)
cv_new_review=cv.transform([processed_new_review]).toarray()
prediction=IMDB.predict(cv_new_review)
```

```python
print(f'Movie review: {'positive' if prediction == 1 else 'Negative'}')
```

```python
import pandas as pd
tf=pd.read_csv('/content/twitter.csv')
```

```python
tf.head()
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
ts = PorterStemmer()
```

```python
def preprocess_text(text):
  words= word_tokenize(text.lower())
  tw=stopwords.words('english')
  words=[i for i in words if i.isalnum() and i not in tw]
  words=[ts.stem(i) for i in words]
  return ' '.join(words)
```

```python
tf['preprocessed_review']=tf['tweet'].apply(preprocess_text)
```

```python
tf.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
tf = tf.drop('id', axis=1)
```

```python
tf.head()
```

```python
tf.isnull().sum()
```

```python
tf['label'].value_counts()
```

```python
tf = tf.dropna().reset_index(drop=True)
```

```python
Cv= CountVectorizer(max_features=5000)
```

```python
X = Cv.fit_transform(tf['preprocessed_review']).toarray()
```

```python
Y = tf['label']
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
Tweet=MultinomialNB()
Tweet
```

```python
Tweet.fit(Xtrain,Ytrain)
```

```python
Ypred=Tweet.predict(Xtest)
Ypred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
Ac=accuracy_score(Ytest,Ypred)
Ac
```

```python
Report=classification_report(Ytest,Ypred)
Report
```
